# 새로운 데이터셋
Seaborn.tips 데이터 셋

```python
import seaborn as sns
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
# 1변량 그래프 : 히스토그램
```

```python
plt.hist(data=tips_df, x='total_bill', bins=20)
```

```python
fig = plt.figure(figsize=(12,5))
axes1 = fig.add_subplot(1,2,1)  #total_bill
axes2 = fig.add_subplot(1,2,2) #tips
axes1.hist(data=tips_df, x='total_bill', bins=20)
axes2.hist(data=tips_df, x='tip', bins=20)
plt.tight_layout()
plt.show()
```

```python
#boxplot
fig = plt.figure(figsize=(12,5))
axes = fig.add_subplot(1,1,1)  #total_bill
axes.boxplot(data=tips_df, x='total_bill')
plt.show()
```

```python
fig = plt.figure(figsize=(12,5))
axes = fig.add_subplot(1,1,1)  #total_bill
axes.boxplot(
  x=[
    tips_df[tips_df['sex']=='Female']['total_bill'],
    tips_df[tips_df['sex']=='Male']['total_bill']
  ],
  labels=['Female','Male']
)
plt.show()
```

```python
fig = plt.figure(figsize=[10,5])
axes = fig.add_subplot(1,1,1)
axes.boxplot(x=
             [tips_df[(tips_df['day'] == 'Sat') & (tips_df['time'] == 'Lunch')]['tip'],
              tips_df[(tips_df['day'] == 'Sat') & (tips_df['time'] == 'Dinner')]['tip']],
             labels=["Lunch","Dinner"])
plt.show()
```

```python
tips_df.groupby(['day','time']).size()
```

```python
tips_df[(tips_df['day'] == 'Sat') & (tips_df['time'] == 'Lunch')]
```

```python
# scatter, 산점도
fig = plt.figure()
axes = fig.add_subplot(1,1,1)  #total_bill
axes.scatter(
  tips_df['total_bill'], tips_df['tip']
  )
plt.show()

```

```python
colors = {"Female":"red", "Male":"blue"}

# 다변량 표현
#성별에 따라 - 색상 
#값의 크기에 따라 - 점의 크기
fig = plt.figure()
axes = fig.add_subplot(1,1,1)  #total_bill
axes.scatter(
  tips_df['total_bill'], tips_df['tip'],
  s=tips_df['size']**2*10,  #tablesize, tip
  c=tips_df['sex'].map(colors),
  alpha=0.5
  )
plt.show()
```

```python
#colors = {"Female":"red", "Male":"blue"}
colors = {"Female":"#ff00ad", "Male":"#000080"} #16진색상코드
# 다변량 표현
#성별에 따라 - 색상 
#값의 크기에 따라 - 점의 크기
fig = plt.figure()
axes = fig.add_subplot(1,1,1)  #total_bill
axes.scatter(
  tips_df['total_bill'], tips_df['tip'],
  s=tips_df['size']**2*10,  #tablesize, tip
  c=tips_df['sex'].map(colors),
  alpha=0.5
  )
plt.show()
```

# 다변량 표현
- 성별에 따라 : 색상 
- 값의 크기에 따라 : 점의 크기

```python
#colors = {"Female":"red", "Male":"blue"}
colors = {"Female":"#ff00ad", "Male":"#000080"} #16진색상코드
tips_df['tip_pct'] = tips_df['tip']/tips_df['total_bill']

fig = plt.figure()
axes = fig.add_subplot(1,1,1)  #total_bill
axes.scatter(
  tips_df['total_bill'], tips_df['tip'],
  s=tips_df['tip_pct']*1000,  #tablesize, tip
  c=tips_df['sex'].map(colors),
  alpha=0.5
  )
plt.show()
```
